# Data cleaning and preprocessing

```python
df_cleaned=df.copy()
#a new variable is created that copies the values from the above
```

```python
df_cleaned.shape
```

```python
df_cleaned.drop_duplicates(inplace=True)
#removes the duplicated values
```

```python
df_cleaned.shape
```

```python
#for ML models all the strings value must be converted into numeric values through binary encoding or one hot encoding
#here in this dataset that kind of values are sex,region,smoker
```

```python
df_cleaned['sex'].value_counts() 
#gives no of values that are similar
#from this info there are only two kinds 0f values in sex columns either male or female
#so male=1 and female= 0 through binary encoding
```

```python
df_cleaned['sex']=df_cleaned['sex'].map({"male":1, "female":0})
```

```python
df_cleaned['smoker'].value_counts()
```

```python
df_cleaned['smoker']=df_cleaned['smoker'].map({"yes":1, "no":0})
```

```python
df_cleaned.head()
```

```python
#df_cleaned['region'].value_counts()
```

```python
#df_cleaned['region']=df_cleaned['region'].map({"southeast":0,"southwest":1,"northwest":2,"northeast":3})
#this is the normal mapping/binary mapping
#many kind of anamolies can occur in this type of encoding so another type is preffered
```

```python
df_cleaned =pd.get_dummies(df_cleaned,columns = ['region'] ,drop_first=True)
```

```python
df_cleaned=df_cleaned.astype(int)
#convert true and false into 0 and 1
```

```python
df_cleaned.head()
```

# Feature Engineering

```python
#PURPOSE: CREATE NEW FEATURE(COLUMNS) FROM THE EXISTING ONE
#Helps to increase the accuracy
```

```python
#i can now classify age into kid(1-10) youth(10-18),adult(18-60),old(60-100)
df_cleaned['age_category']=pd.cut( #cut function helps to create bins from the continuos data
    df_cleaned['age'],#data
    bins=[1,10,18,60,100],#dividing into category
    labels=["kid","youth","adult","old"]
)
```

```python
df_cleaned.head()
```

```python
df_cleaned =pd.get_dummies(df_cleaned,columns = ['age_category'] ,drop_first=True)
#one hot encoding for the newly added category.
```

```python
df_cleaned=df_cleaned.astype(int)
#data cleaning after adding new cells.
```

```python
df_cleaned.head()
```

```python
from sklearn.preprocessing import StandardScaler
columns=['age','bmi','children']
scaler=StandardScaler()
df_cleaned[columns]=scaler.fit_transform(df_cleaned[columns])
```

```python
df_cleaned.head()
```

```python
df_cleaned.info()
```

```python
print(df_cleaned.columns)
```

```python
cols = ['age', 'sex', 'bmi', 'children', 'smoker', 'charges',
        'region_northwest', 'region_southeast', 'region_southwest',
        'age_category_youth', 'age_category_adult', 'age_category_old']

# Compute correlation of each column with 'charges'
correlations = df_cleaned[cols].corr()['charges'].drop('charges')

# Sort from highest to lowest
sorted_corr = correlations.sort_values(ascending=False)

print(sorted_corr)
#this is the pearson correlation method
#However there are other method like spearman rank correlation and kendall tau correlation.
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Bin the continuous 'charges' column into 3 groups (Low, Medium, High)
df_cleaned['charges_bin'] = pd.qcut(df_cleaned['charges'], q=3, labels=['Low', 'Medium', 'High'])

# Step 2: Define categorical columns
categorical_cols = ['sex', 'smoker', 'region_northwest', 'region_southeast',
                    'region_southwest', 'age_category_youth',
                    'age_category_adult', 'age_category_old']

# Step 3: Perform Chi-square test
results = []
for col in categorical_cols:
    contingency_table = pd.crosstab(df_cleaned[col], df_cleaned['charges_bin'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    decision = "Keep Feature" if p < 0.05 else "Remove Feature"
    results.append([col, chi2, p, decision])

# Step 4: Convert to DataFrame and sort by Chi-square value
chi_results_df = pd.DataFrame(results, columns=['Column', 'Chi-square Value', 'p-value', 'Decision'])
chi_results_df = chi_results_df.sort_values(by='Chi-square Value', ascending=False)

print(chi_results_df)
```

```python
final_df=df_cleaned[['age', 'smoker', 'charges','bmi']]
```

PHASE 2

```python
from sklearn.model_selection import train_test_split
```

```python
X=final_df.drop('charges',axis=1)
y=final_df['charges']
#x=all data except charges.
#y=charges
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
#splitting the dataset into 80% training and 20% testing data.
```

```python
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
#to check if data is splitted accordingly.
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# create a model instance
model = LinearRegression()

# fit the model on training data
model.fit(X_train, y_train)
```

```python
# make predictions on test data
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
n=X_test.shape[0]
p=X_test.shape[1]
adjusted_r2=1-((1-r2)*(n-1)/(n-p-1))
adjusted_r2
```
