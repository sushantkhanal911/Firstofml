```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import warnings as warning
warning.filterwarnings('ignore')
```

```python
df=sns.load_dataset("titanic")
```

```python
df.head()
```

# EDA

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop('deck',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df.dropna(subset=['embarked'],inplace = True)
```

```python
df.drop(['class','who','adult_male','alive','embark_town'],axis=1,inplace=True)
#Dropping all the attributes we dont need
#axis=1 for columns.
```

```python
df.head()
```

# Data preprocessing


```python
from sklearn.preprocessing import LabelEncoder
catgorical_cols=['sex','embarked','alone']
label_encoders={}
for col in catgorical_cols:
    le=LabelEncoder()
    df[col]=le.fit_transform(df[col])
```

```python
df=df.astype(int)
```

```python
df.head()
```

# Model

```python
X=df.drop(['survived'],axis=1)
Y=df['survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
accuracy_score(Y_pred,Y_test)
```

```python
confusion_matrix(Y_pred,Y_test)
```

```python
print(classification_report(Y_pred,Y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
naive_model=GaussianNB()
naive_model.fit(X_train,Y_train)
```

```python
y_prednaive=naive_model.predict(X_test)
```

```python
accuracy_score(Y_test,y_prednaive)
```

```python
confusion_matrix(Y_test,y_prednaive)
```

```python
print(classification_report(y_prednaive,Y_test))
```
